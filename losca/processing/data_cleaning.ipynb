```python
# Function to detect and replace outliers with median
def replace_outliers_with_median(df):
    df = df.copy()  # Avoid mutating the original dataframe
    datetime_columns = df.select_dtypes(include=['datetimetz']).columns
    for column in df.select_dtypes(include=[np.number]).columns:
        if column not in datetime_columns:
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            # Replace outliers with median
            median = df[column].median()
            df[column + 'outlier'] = (df[column] < lower_bound) | (df[column] > upper_bound)
            df[column] = np.where((df[column] < lower_bound) | (df[column] > upper_bound), median, df[column])
            
        
    return df

# Function to detect and replace outliers with mean
def replace_outliers_with_mean(df):
    df = df.copy()  # Avoid mutating the original dataframe
    datetime_columns = df.select_dtypes(include=['datetimetz']).columns
    for column in df.select_dtypes(include=[np.number]).columns:
        if column not in datetime_columns:
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            # Replace outliers with mean, add new column with boolean values if the value is an outlier
            mean = df[column].mean()
            df[column + 'outlier'] = ((df[column] < lower_bound) | (df[column] > upper_bound))
            df[column] = np.where((df[column] < lower_bound) | (df[column] > upper_bound), mean, df[column])
            
    
    return df
```

```python
# Function to detect and replace outliers with KNN estimator

from sklearn.impute import KNNImputer

def replace_outliers_with_knn(df):

    df = df.copy()  # Avoid mutating the original dataframe
    datetime_columns = df.select_dtypes(include=['datetimetz']).columns
    for column in df.select_dtypes(include=[np.number]).columns:
        if column not in datetime_columns:
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            # Replace outliers with KNN estimator, add new column with boolean values if the value is an outlier
            df[column + 'outlier'] = ((df[column] < lower_bound) | (df[column] > upper_bound))
            imputer = KNNImputer(n_neighbors=2)
            df[column] = np.where((df[column] < lower_bound) | (df[column] > upper_bound), imputer.fit_transform(df[column].values.reshape(-1, 1)), df[column])
            
    return df

```

```python
mean_imputation_data = replace_outliers_with_mean(unprocessed_data_drop)
median_imputation_data = replace_outliers_with_median(unprocessed_data_drop)
```

```python

knn_imputation_data = replace_outliers_with_knn(unproccesed_data_minusone)
```
